```python
from os import listdir
import pandas as pd
import numpy as np
```

```python
a = listdir(r'D:\Git\Estudos\Jrpgs\Dados')
```

```python
b = a[1:]
```

```python
len(b)
```

```python
caminho = 'Dados/'+a[0]
df = pd.read_csv(caminho, sep=';')
df
```

```python
for i in b:
    caminho = 'Dados/'+b[0]
    df1 = pd.read_csv(caminho, sep=';')
    df = pd.concat([df, df1])
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df['Main Story'] = df['Main Story'].str.replace(' ', '')
df['Main + Extras'] = df['Main + Extras'].str.replace(' ', '')
df['Completionist'] = df['Completionist'].str.replace(' ', '')
df['All Styles'] = df['All Styles'].str.replace(' ', '')
df.head()
```

```python
df = df.replace("NAN", np.nan)
df.head()
```

```python
df['Main Story'] = df['Main Story'].replace('--', np.nan)
df['Main + Extras'] = df['Main + Extras'].replace('--', np.nan)
df['Completionist'] = df['Completionist'].replace('--', np.nan)
df['All Styles'] = df['All Styles'].replace('--', np.nan)
df.head()
```
